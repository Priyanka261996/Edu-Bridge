# WHO Air Quality Database 2018

```python
Air pollution is a major environmental risk to health. By reducing air pollution levels, countries can reduce the burden of disease from stroke, heart disease, lung cancer, and both chronic and acute respiratory diseases, including asthma.
The lower the levels of air pollution, the better the cardiovascular and respiratory health of the population will be, both long- and short-term.
Particulate matter (PM) is a common proxy indicator for air pollution. It affects more people than any other pollutant.
WHO Air quality guideline values
Particulate matter (PM)
Guideline values
Fine particulate matter (PM2.5)
10 μg/m3 annual mean
25 μg/m3 24-hour mean

Coarse particulate matter (PM10)
20 μg/m3 annual mean
50 μg/m3 24-hour mean
```

# To create Scatterplot

```python

sns.scatterplot(x=d['Region'],y=d['pm10'])
plt.title('Region vs pm25,size=16')
plt.ylabel('pm10',size=16)
plt.xlabel('Region',size=15)
plt.show()
```

# To create Bar graph

```python

plt.bar(d["Year"],d["pm10"])
plt.title('ID WHO city VS pm10',fontsize=15,color='red')
plt.xlabel('ID WHO city',fontsize=15)
plt.ylabel('pm10',fontsize=15)
plt.grid(False)
plt.show()
```

```python
# Find : In year 2010 , concentration of pm10 is higher.
```

```python

plt.bar(d["Year"],d["pm25"])
plt.title('ID WHO city VS pm25',fontsize=15,color='red')
plt.xlabel('ID WHO city',fontsize=15)
plt.ylabel('pm25',fontsize=15)
plt.grid(False)
plt.show()
```

```python
# Find : In year 2013 , concentration of pm25 is higher.
```
